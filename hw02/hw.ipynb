```python
print("\nколичество пропусков")
missing_values = df.isnull().sum().sort_values(ascending=False)
print(missing_values[missing_values > 0])

df_model = df.copy()

#заполнение пропусков медианой и мода, только для eda, для моделирования имутеры на train данных
from sklearn.impute import SimpleImputer
if df.isnull().sum().sum() > 0: 
    
    imputer = SimpleImputer(strategy='median')
    numeric_cols = df_model.select_dtypes(include=['number']).columns
    df_model[numeric_cols] = imputer.fit_transform(df_model[numeric_cols])
    print("числовые пропуски заполнены медианой")

    for col in df_model.select_dtypes(include=['object', 'str']).columns:
        mode_value = df_model[col].mode()[0] 
        df_model[col] = df_model[col].fillna(mode_value)
    
else:
    print("пропусков нет")
```

```python
#используем для порядковых признаков
from sklearn.preprocessing import LabelEncoder

le_grade = LabelEncoder()
df_model['Grade_encoded'] = le_grade.fit_transform(df_model['Grade'])
grade_order = ['F', 'D', 'C', 'B', 'A']
grade_mapping = {grade: idx for idx, grade in enumerate(grade_order)}
df_model['Grade_ordinal'] = df_model['Grade'].map(grade_mapping)
print(f"кодирование Grade: {dict(zip(le_grade.classes_, le_grade.transform(le_grade.classes_)))}")
print(f"Ordinal: {grade_mapping}")

cat_cols = ['Gender', 'Department', 'Parent_Education_Level', 'Family_Income_Level']
df_model = pd.get_dummies(df_model, columns=cat_cols, drop_first=True)

df_model = pd.get_dummies(df_model, 
                          columns=['Extracurricular_Activities', 'Internet_Access_at_Home'], 
                          drop_first=True)

print(f"после OneHot: {df_model.shape[1]} признаков (было {df.shape[1]})")
display(df_model.head())
```

```python
df_model['Study_Hours_log'] = np.log1p(df['Study_Hours_per_Week'])
```

```python
print("\nновые признаки")
#отношение заданий к тестам
if 'Assignments_Avg' in df.columns and 'Quizzes_Avg' in df.columns:
    df_model['assignment_quiz_ratio'] = (df_model['Assignments_Avg'] + 1) / (df_model['Quizzes_Avg'] + 1)

#учёба + стресс
if 'Study_Hours_per_Week' in df.columns and 'Stress_Level (1-10)' in df.columns:
    df_model['high_load_flag'] = ((df_model['Study_Hours_per_Week'] > 20) & 
                           (df_model['Stress_Level (1-10)'] >= 7)).astype(int)

#редкие оценки
if 'Grade' in df.columns:
    df_model['Grade_grouped'] = df_model['Grade'].apply(lambda x: 'extreme' if x in ['A', 'F'] else 'middle')
```

# итоговый отчёт по EDA: students performance

# что я поняла про датасет:

1. Размер данных: 5000 студентов, 23 признака
2. Целевая переменная: `Grade` (A-F), сильный дисбаланс: ~46% — оценка C
3. Пропуски: только в `Parent_Education_Level` (1025 пропусков, ~20%)
4. Числовые признаки: оценки (0-100), часы учёбы, стресс, сон
5. Категориальные: пол, факультет, доход семьи, доступ к интернету
6. Сильные корреляции: `Projects_Score` и `Final_Score` с `Total_Score` (>0.5)
7. Распределение `Total_Score` скошено вправо — большинство студентов имеют средние баллы
8. Выбросы есть в `Study_Hours_per_Week` (некоторые учатся >40 часов/неделю)
9. `Attendance (%)` положительно связан с успеваемостью
10. Стресс (`Stress_Level`) слабо коррелирует с оценками — не всегда мешает учёбе

# гипотезы и наблюдения:

1. студенты с высшим образованием у родителей чаще получают высокие оценки
2. доступ к интернету дома положительно влияет на успеваемость (гипотеза для проверки)
3. студенты с оценкой "A" спят в среднем больше, чем с оценкой "F"
