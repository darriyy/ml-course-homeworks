```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

print("download dataset")
df = pd.read_csv('Students Performance Dataset.csv')
```

```python
print("\nобщая информация")
df.info()
```

```python
print("\nописание датасета")
df.describe()
```

```python
print("\nколичество объектов каждого класса")
print(df['Grade'].value_counts())
print("\nбаланс классов")
print(df['Grade'].value_counts(normalize=True) * 100)
```

```python
print("\nколичество пропусков")
missing_values = df.isnull().sum().sort_values(ascending=False)
print(missing_values[missing_values > 0])

#заполнение пропусков медиантой, тк если они появятся, их будет мало
from sklearn.impute import SimpleImputer
if df.isnull().sum().sum() > 0: 
    imputer = SimpleImputer(strategy='median')
    numeric_cols = df.select_dtypes(include=['number']).columns

    df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
    print("пропуски заполнены медианой")
else:
    print("пропусков нет")
```

```python
print(f"\nпроверка на дубликаты {df.duplicated().sum()}")

print("\nкатегориальные признаки")
for col in ['Gender', 'Department', 'Parent_Education_Level', 'Family_Income_Level']:
    if col in df.columns:
        print(f"{col}: {df[col].nunique()}")
```
